```python
import pandas as pd
import os
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
features = pd.read_csv(os.path.join("/home/mmh/new/flaskDashboard/data/rawData", "featureData.csv"))
features = features.rename(columns = {"participantId": "id"})
features.head()
```

```python
labels = pd.read_csv(os.path.join("/home/mmh/gitlab/analysis/data", "mentalHealthIndicators.csv"))
labels = labels.rename(columns = {"record_id": "id", "dass":"label1", "current_diagnosis":"label2"})
labels.drop("phone", axis=1, inplace=True)
labels.head()
```

```python
labels.id.unique().tolist()
```

```python
df = pd.merge(left=features, right=labels, how="inner", on="id")
df["max_screen_on_time"] = round(df["max_screen_on_time"]/60, 2)
df["total_screen_on_time"] = round(df["total_screen_on_time"]/60, 2)
df["min_duration_of_single_incoming_call"] = round(df["min_duration_of_single_incoming_call"]/60, 2)
df["min_duration_of_single_outgoing_call"] = round(df["min_duration_of_single_outgoing_call"]/60, 2)
df["max_duration_of_single_incoming_call"] = round(df["max_duration_of_single_incoming_call"]/60, 2)
df["max_duration_of_single_outgoing_call"] = round(df["max_duration_of_single_outgoing_call"]/60, 2)

df["total_duration_of_incoming_calls"] = round(df["total_duration_of_incoming_calls"]/60, 2)
df["total_duration_of_outgoing_calls"] = round(df["total_duration_of_outgoing_calls"]/60, 2)
df["total_duration_of_all_calls"] = round(df["total_duration_of_all_calls"]/60, 2)
df.head()
```

```python
df.id.unique()
```

```python
df_agg = df.groupby("id").mean().round(2).reset_index()
df_agg.drop("date_num", axis=1, inplace=True)
df_agg["no_of_unlocks"] = df_agg["no_of_unlocks"].apply(np.ceil)
df_agg["no_of_missed_calls"] = df_agg["no_of_missed_calls"].apply(np.ceil)
df_agg["no_of_rejected_calls"] = df_agg["no_of_rejected_calls"].apply(np.ceil)
df_agg["no_of_incoming_calls"] = df_agg["no_of_incoming_calls"].apply(np.ceil)
df_agg["no_of_outgoing_calls"] = df_agg["no_of_outgoing_calls"].apply(np.ceil)
df_agg["total_no_of_calls"] = df_agg["total_no_of_calls"].apply(np.ceil)
df_agg.head()
```
