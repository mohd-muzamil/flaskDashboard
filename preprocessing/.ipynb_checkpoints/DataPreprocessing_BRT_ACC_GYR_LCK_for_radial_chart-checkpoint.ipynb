```python
import pandas as pd
import os
from pathlib import Path
import numpy as np
import datetime
pd.set_option('display.max_rows', None)

dataPath1 = "/csv/backup_frigg1"
dataPath2 = "/csv/backup"

output_path = Path("/home/mmh/flaskDashboard/data/processedData")

# input file names
brightnessIosFilename = "Brightness.csv"

accelerometerIosFilename = "Accelerometer.csv"

gyroscopeIosFilename = "Gyroscope.csv"

lockstateIosFilename = "Lock_state.csv"


# output file names
brightnessFilename = "brightness_processed.csv"
accelerometerFilename = "accelerometer_processed.csv"
gyroscopeFilename = "gyroscope_processed.csv"
lockstateFilename = "lockstate_processed.csv"

# Filtered participants from another notebook
filteredParticipants = ['PROSITC1005', 'PROSITC1014', 'PROSITC1018', 'PROSITC1025', 'PROSITC1026', 'PROSITC1027', 'PROSITC1029', 'PROSITC1030', 'PROSITC1031', 'PROSITC1033', 'PROSITC1034', 'PROSITC1036', 'PROSITC1037', 'PROSITC1038', 'PROSITC1039', 'PROSITC1040', 'PROSITC1041', 'PROSITC1043', 'PROSITC1044', 'PROSITC1046', 'PROSITC1057', 'PROSITC1061', 'PROSITC1062', 'PROSITC1066', 'PROSITC1070', 'PROSITC1079', 'PROSITC1080', 'PROSITC1082', 'PROSITC1084', 'PROSITC1086', 'PROSITC1089', 'PROSITC1090', 'PROSITC1097', 'PROSITC1100', 'PROSITC1101', 'PROSITC1104', 'PROSITC1109', 'PROSITC1110', 'PROSITC1111', 'PROSITC1119', 'PROSITC1122', 'PROSITC1126', 'PROSITC1128', 'PROSITC1130', 'PROSITC1132', 'PROSITC1135', 'PROSITC1136', 'PROSITC1137', 'PROSITC1140', 'PROSITC1153', 'PROSITC1158', 'PROSITC1174', 'PROSITC1187', 'PROSITC1192', 'PROSITC1193', 'PROSITC1200', 'PROSITC1221', 'PROSITC1222', 'PROSITC1223', 'PROSITC1224', 'PROSITC1227', 'PROSITC1237', 'PROSITC1240', 'PROSITC1244', 'PROSITC1261', 'PROSITC1272', 'PROSITC1293', 'PROSITC1297', 'PROSITC1299', 'PROSITC1304', 'PROSITC1308', 'PROSITC1317', 'PROSITC1325', 'PROSITC1326', 'PROSITC1343', 'PROSITC1369', 'PROSITC1372', 'PROSITC1379', 'PROSITC1412', 'PROSITC1413', 'PROSITC1422', 'PROSITC1424', 'PROSITC1428', 'PROSITC1438', 'PROSITC1493', 'PROSITC1501', 'PROSITC1502', 'PROSITC1512', 'PROSITC1516', 'PROSITC1526', 'PROSITC1529', 'PROSITC1548', 'PROSITC1550', 'PROSITC1553', 'PROSITC1555', 'PROSITC1562', 'PROSITC1564', 'PROSITC1570', 'PROSITC1602', 'PROSITC1615', 'PROSITC1621', 'PROSITC1623', 'PROSITC1625', 'PROSITC1627', 'PROSITC1629', 'PROSITC1635', 'PROSITC1639', 'PROSITC1642', 'PROSITC1663', 'PROSITC1666', 'PROSITC1671', 'PROSITC1672', 'PROSITC1678', 'PROSITC1684', 'PROSITC1685', 'PROSITC1705', 'PROSITC1705', 'PROSITC1712', 'PROSITC1722', 'PROSITC1748', 'PROSITC1754', 'PROSITC1756', 'PROSITC1761', 'PROSITC1762', 'PROSITC1764', 'PROSITC1785', 'PROSITC1795', 'PROSITC1798', 'PROSITC1803', 'PROSITC1807', 'PROSITC1809', 'PROSITC1810', 'PROSITC1823', 'PROSITC1827', 'PROSITC1831', 'PROSITC1835', 'PROSITC1844', 'PROSITC1861', 'PROSITC1863', 'PROSITC1864', 'PROSITC1865', 'PROSITC1874', 'PROSITC1877', 'PROSITC1881', 'PROSITC1895', 'PROSITC1898', 'PROSITC1911', 'PROSITC1914', 'PROSITC1926', 'PROSITC1942', 'PROSITC1958', 'PROSITC1966', 'PROSITC1979', 'PROSITC1980', 'PROSITC1983', 'PROSITC1985', 'PROSITC1996', 'PROSITC1999', 'PROSITC2017', 'PROSITC2029', 'PROSITC2059', 'PROSITC2060', 'PROSITC2065', 'PROSITC2069', 'PROSITC2076', 'PROSITC2096', 'PROSITC2111', 'PROSITC2117', 'PROSITC2129', 'PROSITC2134', 'PROSITC2156', 'PROSITC2162', 'PROSITC2170', 'PROSITC2195', 'PROSITC2202', 'PROSITC2203', 'PROSITC2135', 'PROSITC2207', 'PROSITC2233', 'PROSITC2259', 'PROSITC2264', 'PROSITC2265', 'PROSITC2267', 'PROSITC2269', 'PROSITC2271', 'PROSITC2278', 'PROSITC2281', 'PROSITC2303', 'PROSITC2306', 'PROSITC2315', 'PROSITC2323', 'PROSITC2327', 'PROSITC2330', 'PROSITC2332', 'PROSITC2333', 'PROSITC2360', 'PROSITC2382', 'PROSITC2384', 'PROSITC2394', 'PROSITC2396', 'PROSITC2411', 'PROSITC2928', 'PROSITC2964', 'PROSITC2975', 'PROSITC2977', 'PROSITC2984', 'PROSITC2986', 'PROSITC3001', 'PROSITC3014', 'PROSITC3025', 'PROSITC3029', 'PROSITC3031', 'PROSITC3035', 'PROSITC3037', 'PROSITC3038', 'PROSITC3044', 'PROSITC3053', 'PROSITC3067', 'PROSITC3070', 'PROSITC3071', 'PROSITC3073', 'PROSITC3076', 'PROSITC3078', 'PROSITC3079', 'PROSITC3102', 'PROSITC3110', 'PROSITC3114', 'PROSITC3117', 'PROSITC3121', 'PROSITC3134', 'PROSITC3150']

### Brightness
#IOS data
header_list = ["id", "participant", "attribute", "brt", "timestamp", "uploadtimestamp"]
brightness_ios1 = pd.read_csv(os.path.join(dataPath1, brightnessIosFilename), sep='|', header=None, names=header_list)
brightness_ios2 = pd.read_csv(os.path.join(dataPath2, brightnessIosFilename), sep='|', header=None, names=header_list)
# brightness_ios["device"] = "ios"

# Android data
# brightness_android = pd.read_csv(os.path.join(data_path, brightnessAndroidFilename), sep='|', header=None, names=header_list)
# brightness_android["device"] = "android"

# decided not to normalize the data since it was converting most values to 0
#normalizing the brightnesLevel
# brightness_android["brt"] = ((brightness_android["brt"] - brightness_android["brt"].min()) / (brightness_android["brt"].max() - brightness_android["brt"].min())).round(1)

#Combine both datasets
# brightness = pd.concat([brightness_ios, brightness_android], ignore_index=True)
brightness = pd.concat([brightness_ios1, brightness_ios2], ignore_index=True)

brightness.drop_duplicates(inplace=True)
brightness["timestamp"] = pd.to_datetime(brightness["timestamp"], utc=False)

#sort values
brightness.sort_values(["participant", "timestamp"], inplace=True, ascending=True)
brightness = brightness.reset_index(drop=True)

#add necessary columns
brightness["date"] = brightness.timestamp.dt.date
brightness["minuteOfTheDay"] = (brightness.timestamp - brightness.timestamp.dt.floor('d')).astype('timedelta64[m]').astype('int')

#keep only necessary columns
# brightness = brightness[["participant", "device", "date", "minuteOfTheDay", "brt"]]
brightness = brightness[["participant", "date", "minuteOfTheDay", "brt"]]

# keep only filtered participants
brightness = brightness[brightness["participant"].isin(filteredParticipants)]

#Aggregate the data over a period of 1minute
# brightnessAgg = brightness.groupby(["participant", "device", "date", "minuteOfTheDay"])["brt"].agg("mean").reset_index()
brightnessAgg = brightness.groupby(["participant", "date", "minuteOfTheDay"])["brt"].agg("mean").reset_index()
brightnessAgg["brt"] = round(brightnessAgg["brt"], 2)

# temp = lockstate.merge(brightnessAgg, on=["participantId", "device", "date", "minuteOfTheDay"], how="inner")
# brightnessAgg = temp[temp.lck==1].copy()
# brightnessAgg.drop("lck", axis=1, inplace=True)

# save the file
brightnessAgg.to_csv(os.path.join(output_path, brightnessFilename), header=True, index=False)

### Accelerometer
#IOS data
header_list = ["id", "participant", "attribute", "accX", "accY", "accZ", "timestamp", "uploadtimestamp"]
accelerometer_ios1 = pd.read_csv(os.path.join(dataPath1, accelerometerIosFilename), sep='|', header=None, names=header_list)
accelerometer_ios2 = pd.read_csv(os.path.join(dataPath2, accelerometerIosFilename), sep='|', header=None, names=header_list)

# accelerometer_ios["device"] = "ios"

# #Android data
# accelerometer_android = pd.read_csv(os.path.join(data_path, accelerometerAndroidFilename), sep='|', header=None, names=header_list)
# accelerometer_android["device"] = "android"

#Combine both datasets
# accelerometer = pd.concat([accelerometer_ios, accelerometer_android], ignore_index=True)
accelerometer = pd.concat([accelerometer_ios1, accelerometer_ios2], ignore_index=True)

accelerometer.drop_duplicates(inplace=True)
accelerometer["timestamp"] = pd.to_datetime(accelerometer["timestamp"], utc=False)

#sort values
accelerometer.sort_values(["participant", "timestamp"], inplace=True, ascending=True)
accelerometer = accelerometer.reset_index(drop=True)

#add necessary columns
accelerometer["date"] = accelerometer.timestamp.dt.date
accelerometer["minuteOfTheDay"] = (accelerometer.timestamp - accelerometer.timestamp.dt.floor('d')).astype('timedelta64[m]').astype('int')
accelerometer["acc"] = np.sqrt(accelerometer.accX**2 + accelerometer.accY**2 + accelerometer.accZ**2)

#keep only necessary columns
# accelerometer = accelerometer[["participant", "device", "date", "minuteOfTheDay", "acc"]]
accelerometer = accelerometer[["participant", "date", "minuteOfTheDay", "acc"]]

# keep only filtered participants
accelerometer = accelerometer[accelerometer["participant"].isin(filteredParticipants)]

#Aggregate the data over a period of 1minute
# accelerometerAgg = accelerometer.groupby(["participant", "device", "date", "minuteOfTheDay"])["acc"].agg("mean").reset_index()
accelerometerAgg = accelerometer.groupby(["participant", "date", "minuteOfTheDay"])["acc"].agg("mean").reset_index()
accelerometerAgg.acc = round(accelerometerAgg.acc, 2)

# save the file
accelerometerAgg.to_csv(os.path.join(output_path, accelerometerFilename), header=True, index=False)

### Gyroscope
#IOS data
header_list = ["id", "participant", "attribute", "gyroX", "gyroY", "gyroZ", "timestamp", "uploadtimestamp"]
gyroscope_ios1 = pd.read_csv(os.path.join(dataPath1, gyroscopeIosFilename), sep='|', header=None, names=header_list)
gyroscope_ios2 = pd.read_csv(os.path.join(dataPath2, gyroscopeIosFilename), sep='|', header=None, names=header_list)
# gyroscope_ios["device"] = "ios"

# # Android data
# gyroscope_android = pd.read_csv(os.path.join(data_path, gyroscopeAndroidFilename), sep='|', header=None, names=header_list)
# gyroscope_android["device"] = "android"

#Combine both datasets
# gyroscope = pd.concat([gyroscope_ios, gyroscope_android], ignore_index=True)
gyroscope = pd.concat([gyroscope_ios1, gyroscope_ios2], ignore_index=True)


gyroscope.drop_duplicates(inplace=True)
gyroscope["timestamp"] = pd.to_datetime(gyroscope["timestamp"], utc=False)

#sort values
gyroscope.sort_values(["participant", "timestamp"], inplace=True, ascending=True)
gyroscope = gyroscope.reset_index(drop=True)

#add necessary columns
gyroscope["date"] = gyroscope.timestamp.dt.date
gyroscope["minuteOfTheDay"] = (gyroscope.timestamp - gyroscope.timestamp.dt.floor('d')).astype('timedelta64[m]').astype('int')
gyroscope["gyr"] = np.sqrt(gyroscope.gyroX**2 + gyroscope.gyroY**2 + gyroscope.gyroZ**2)

#keep only necessary columns
# gyroscope = gyroscope[["participant", "device", "date", "minuteOfTheDay", "gyr"]]
gyroscope = gyroscope[["participant", "date", "minuteOfTheDay", "gyr"]]

# keep only filtered participants
gyroscope = gyroscope[gyroscope["participant"].isin(filteredParticipants)]

#Aggregate the data over a period of 1minute
# gyroscopeAgg = gyroscope.groupby(["participant", "device", "date", "minuteOfTheDay"])["gyr"].agg("mean").reset_index()
gyroscopeAgg = gyroscope.groupby(["participant", "date", "minuteOfTheDay"])["gyr"].agg("mean").reset_index()
gyroscopeAgg.gyr = round(gyroscopeAgg.gyr, 2)

# save the file
gyroscopeAgg.to_csv(os.path.join(output_path, gyroscopeFilename), header=True, index=False)

### Lockstate
#IOS data
header_list = ["id", "participant", "attribute", "lck", "timestamp", "uploadtimestamp"]
lockstate_ios1 = pd.read_csv(os.path.join(dataPath1, lockstateIosFilename), sep='|', header=None, names=header_list)
lockstate_ios2 = pd.read_csv(os.path.join(dataPath2, lockstateIosFilename), sep='|', header=None, names=header_list)
# lockstate_ios["device"] = "ios"

# #Android data
# lockstate_android = pd.read_csv(os.path.join(data_path, lockstateAndroidFilename), sep='|', header=None, names=header_list)
# lockstate_android["device"] = "android"

#Combine both datasets
# lockstate = pd.concat([lockstate_ios, lockstate_android], ignore_index=True)
lockstate = pd.concat([lockstate_ios1, lockstate_ios2], ignore_index=True)

lockstate.drop_duplicates(inplace=True)
lockstate["timestamp"] = pd.to_datetime(lockstate["timestamp"], utc=False)

#sort values
lockstate.sort_values(["participant", "timestamp"], inplace=True, ascending=True)
lockstate = lockstate.reset_index(drop=True)

#add necessary columns
lockstate["date"] = lockstate.timestamp.dt.date
lockstate["minuteOfTheDay"] = (lockstate.timestamp - lockstate.timestamp.dt.floor('d')).astype('timedelta64[m]').astype('int')

#keep only necessary columns
# lockstate = lockstate[["participantId", "device", "date", "minuteOfTheDay", "lck"]]
lockstate = lockstate[["participantId", "date", "minuteOfTheDay", "lck"]]

# keep only filtered participants
lockstate = lockstate[lockstate["participant"].isin(filteredParticipants)]

lockstate = lockstate[(lockstate.lck == "screen_on") | (lockstate.lck == "screen_off") | (lockstate.lck == "UNLOCKED") | (lockstate.lck == "LOCKED")].copy()
lockstate.loc[lockstate.lck=="screen_on", "lck"] = 1
lockstate.loc[lockstate.lck=="UNLOCKED", "lck"] = 1
lockstate.loc[lockstate.lck=="screen_off", "lck"] = 0
lockstate.loc[lockstate.lck=="LOCKED", "lck"] = 0

# save the file
lockstate.to_csv(os.path.join(output_path, lockstateFilename), header=True, index=False)

print("End")
```
